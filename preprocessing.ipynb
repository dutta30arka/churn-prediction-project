```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('unique.csv')

# Define the threshold (70% in this case)
threshold = 0.70

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Dictionary to store the unique value ratio for each column
unique_ratios = {}

# Iterate over each column
for column in df.columns:
    # Calculate the ratio of unique values to the total number of rows
    unique_ratio = df[column].nunique() / total_rows
    # Store the unique ratio
    unique_ratios[column] = unique_ratio

    # Check if the ratio exceeds the threshold
    if unique_ratio > threshold:
        # Drop the column
        df.drop(column, axis=1, inplace=True)

# Plot the unique value ratios
plt.figure(figsize=(12, 8))
bars = plt.bar(unique_ratios.keys(), unique_ratios.values(), color='skyblue')

# Highlight the bars for columns that are dropped
for bar, ratio in zip(bars, unique_ratios.values()):
    if ratio > threshold:
        bar.set_color('red')

plt.xlabel('Columns')
plt.ylabel('Unique Value Ratio')
plt.title('Unique Value Ratio of Each Column')
plt.xticks(rotation=90)
plt.axhline(y=threshold, color='gray', linestyle='--', linewidth=2)
plt.text(len(unique_ratios)/2, threshold, 'Threshold', ha='center', va='bottom', color='gray')

plt.show()

# Write the modified DataFrame back to a CSV file
df.to_csv('unique_filtered.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv('unique_filtered.csv')

# Define the threshold (0 % in this case)
threshold = 0.0

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Dictionary to store the variance ratio for each column
variance_ratios = {}

# Iterate over each column
for column in df.columns:
    # Calculate the variance of the column
    column_variance = df[column].var()
    # Calculate the ratio of non-zero variance values to the total number of rows
    variance_ratio = column_variance / total_rows
    # Store the variance ratio
    variance_ratios[column] = variance_ratio

    # Check if the ratio is below the threshold
    if variance_ratio < threshold:
        # Drop the column
        df.drop(column, axis=1, inplace=True)

# Plot the variance ratios
plt.figure(figsize=(12, 8))
bars = plt.bar(variance_ratios.keys(), variance_ratios.values(), color='skyblue')

# Highlight the bars for columns that are dropped
for bar, ratio in zip(bars, variance_ratios.values()):
    if ratio < threshold:
        bar.set_color('red')

plt.xlabel('Columns')
plt.ylabel('Variance Ratio')
plt.title('Variance Ratio of Each Column')
plt.xticks(rotation=90)
plt.axhline(y=threshold, color='gray', linestyle='--', linewidth=2)
plt.text(len(variance_ratios)/2, threshold, 'Threshold', ha='center', va='bottom', color='gray')

plt.show()

# Write the modified DataFrame back to a CSV file
df.to_csv('variance_filtered.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('variance_filtered.csv')

# Define the threshold for outlier detection (e.g., 1.5 times the IQR)
threshold = 1.5

# Define a function for outlier treatment using capping and flooring
def treat_outliers(column):
    if pd.api.types.is_numeric_dtype(column):
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR
        # Cap and floor outliers
        column = column.clip(lower=lower_bound, upper=upper_bound)
    return column

# Apply outlier treatment to each column
for column in df.columns:
    df[column] = treat_outliers(df[column])

# Create box plots before and after outlier treatment
plt.figure(figsize=(12, 6))

# Before treatment
plt.subplot(1, 2, 1)
df.boxplot(rot=90)
plt.title('Before Outlier Treatment')
plt.ylabel('Values')

# After treatment
plt.subplot(1, 2, 2)
df.boxplot(rot=90)
plt.title('After Outlier Treatment')
plt.ylabel('Values')

plt.tight_layout()
plt.show()

# Write the modified DataFrame back to a CSV file
df.to_csv('cf_filtered.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('variance_filtered.csv')

# Define a function for outlier treatment using the boxplot method
def treat_outliers(column):
    if pd.api.types.is_numeric_dtype(column):
        Q1 = column.quantile(0.25)
        Q3 = column.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Cap and floor outliers using the boxplot method
        column = column.apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
    return column

# Apply outlier treatment to each column
for column in df.columns:
    df[column] = treat_outliers(df[column])

# Create box plots before and after outlier treatment
plt.figure(figsize=(12, 6))

# Before treatment
plt.subplot(1, 2, 1)
df.boxplot(rot=90)
plt.title('Before Outlier Treatment')
plt.ylabel('Values')

# After treatment
plt.subplot(1, 2, 2)
df.boxplot(rot=90)
plt.title('After Outlier Treatment')
plt.ylabel('Values')

plt.tight_layout()
plt.show()

# Write the modified DataFrame back to a CSV file
df.to_csv('bp_filtered.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('variance_filtered.csv')

# Define the number of standard deviations (sigma) for outlier treatment
sigma = 2

# Define a function for outlier treatment using the standardization sigma approach method
def treat_outliers(column):
    if pd.api.types.is_numeric_dtype(column):
        mean = column.mean()
        std_dev = column.std()
        # Standardize the data
        standardized_values = (column - mean) / std_dev
        # Cap and floor outliers using the standardization sigma approach method
        treated_values = standardized_values.apply(lambda x: sigma if x > sigma else (-sigma if x < -sigma else x))
        # Inverse transform to obtain treated values in the original scale
        treated_column = treated_values * std_dev + mean
    return treated_column

# Apply outlier treatment to each column
for column in df.columns:
    df[column] = treat_outliers(df[column])

# Create box plots before and after outlier treatment
plt.figure(figsize=(12, 6))

# Before treatment
plt.subplot(1, 2, 1)
df.boxplot(rot=90)
plt.title('Before Outlier Treatment')
plt.ylabel('Values')

# After treatment
plt.subplot(1, 2, 2)
df.boxplot(rot=90)
plt.title('After Outlier Treatment')
plt.ylabel('Values')

plt.tight_layout()
plt.show()

# Write the modified DataFrame back to a CSV file
df.to_csv('sa_filtered.csv', index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV dataset
df = pd.read_csv('cf_filtered.csv')

# Define the threshold for correlation
threshold = 0.90

# Calculate the correlation matrix
corr_matrix = df.corr().abs()

# Create a mask to identify highly correlated features
mask = (corr_matrix >= threshold) & (corr_matrix < 1)

# Get the names of highly correlated features
highly_correlated_features = set()
for col in mask.columns:
    correlated_cols = mask.index[mask[col]].tolist()
    if correlated_cols:
        highly_correlated_features.add(col)
        highly_correlated_features.update(correlated_cols)

# Remove highly correlated features from the DataFrame
df_filtered = df.drop(highly_correlated_features, axis=1)

# Plot the correlation matrix before removing highly correlated features
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix Before Removing Highly Correlated Features')
plt.show()

# Calculate the correlation matrix after removing highly correlated features
corr_matrix_filtered = df_filtered.corr().abs()

# Plot the correlation matrix after removing highly correlated features
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_filtered, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix After Removing Highly Correlated Features')
plt.show()

# Write the filtered DataFrame back to a CSV file
df_filtered.to_csv('filtered_dataset_final.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV dataset
df = pd.read_csv('filtered_dataset_final.csv')

# Define the threshold for correlation
threshold = 0.90

# Calculate the correlation matrix
corr_matrix = df.corr().abs()

# Create a mask to identify highly correlated features
mask = (corr_matrix >= threshold) & (corr_matrix < 1)

# Get the names of highly correlated features
highly_correlated_features = set()
for col in mask.columns:
    correlated_cols = mask.index[mask[col]].tolist()
    if correlated_cols:
        highly_correlated_features.add(col)
        highly_correlated_features.update(correlated_cols)

# Remove highly correlated features from the DataFrame
df_filtered = df.drop(highly_correlated_features, axis=1)

# Plot the correlation matrix before removing highly correlated features
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix Before Removing Highly Correlated Features')
plt.show()

# Calculate VIF for remaining features
vif_data = pd.DataFrame()
vif_data["feature"] = df_filtered.columns
vif_data["VIF"] = [variance_inflation_factor(df_filtered.values, i) for i in range(len(df_filtered.columns))]

# Plot the VIF values
plt.figure(figsize=(10, 6))
sns.barplot(x='VIF', y='feature', data=vif_data)
plt.title('Variance Inflation Factor (VIF)')
plt.xlabel('VIF')
plt.ylabel('Features')
plt.show()

# Find features with VIF > 5
high_vif_features = vif_data[vif_data['VIF'] > 5]

# Print features with VIF > 5
print("Features with VIF > 5:")
print(high_vif_features)

# Write the filtered DataFrame back to a CSV file
df_filtered.to_csv('filtered_dataset.csv', index=False)
```
